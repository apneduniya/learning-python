````python
# Example for Question 1: Logical operators
x = 5
y = 10
print(x > 0 and y > 0)  # True
print(x > 0 or y < 0)   # True
print(not(x > 0))       # False
````

````python
## 2. State with an example, the role of the following operators:
### i) `is`  ii) `in`  iii) `not in`  iv) `<<`  v) `>>` (5)

- `is`: Checks if two variables refer to the same object.
- `in`: Checks if a value exists in a sequence.
- `not in`: Checks if a value does not exist in a sequence.
- `<<`: Bitwise left shift operator.
- `>>`: Bitwise right shift operator.

> "The `is` operator checks identity, not equality. `in` and `not in` are used for membership tests. `<<` and `>>` are used for bitwise operations."
````

````python
# Example for Question 2: Special operators

# is operator
a = [1, 2]
b = a
print(a is b)  # True

# in operator
print(3 in [1, 2, 3])  # True

# not in operator
print(4 not in [1, 2, 3])  # True

# << operator (bitwise left shift)
print(2 << 2)  # 8 (2*2^2)

# >> operator (bitwise right shift)
print(8 >> 2)  # 2 (8/2^2)
````

````python
# Question 7: range() function example
print("Range function examples:")
for i in range(1, 6):
    print(i, end=' ')
print()

# Question 8: Odd non-prime numbers between 1 to 50
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

print("Odd non-prime numbers between 1 to 50:")
for i in range(1, 51):
    if i % 2 != 0 and not is_prime(i):
        print(i, end=' ')
print()

# Question 9: Sum of digits
num = int(input("Enter a number for sum of digits: "))
original = num
sum_digits = 0
while num > 0:
    sum_digits += num % 10
    num //= 10
print(f"Sum of digits of {original}:", sum_digits)
````

````python
# Question 10: Number pattern
print("Question 10: Number pattern")
n = 4
for i in range(1, n+1):
    print(str(i) * i)
print()

# Question 11: Series sum calculation
print("Question 11: Series sum")
n = int(input("Enter n for series sum: "))
sum_series = 0
for i in range(1, n+1):
    sum_series += i / (i+1)
print("Sum of series:", sum_series)
print()

# Question 12: break vs continue examples
print("Question 12a: break example")
for i in range(1, 6):
    if i == 3:
        break
    print(i, end=' ')  # Output: 1 2
print()

print("Question 12b: continue example")
for i in range(1, 6):
    if i == 3:
        continue
    print(i, end=' ')  # Output: 1 2 4 5
print()
````

````python
## 20. Define recursion. Differentiate it with iteration. (2+3)

- **Recursion:** When a function calls itself to solve a problem.
- **Difference:**
  - Recursion solves a problem by breaking it into smaller subproblems, calling itself.
  - Iteration uses loops to repeat code until a condition is met.
  - Recursion can be more elegant but may use more memory due to function call stack.

> "Recursion is best for problems that can be divided into similar subproblems."

---

## 21. Write a python program to print Fibonacci series using recursion (5)

---

## 22. Write a python program to print Factorial using recursion (5)

---

## 23. Define a recursive function to find out sum of digits of a number (number should be user input). (5)
````

````python
# Question 21: Fibonacci series using recursion
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)

print("Question 21: Fibonacci series using recursion")
n = int(input("Enter number of terms for Fibonacci: "))
for i in range(n):
    print(fibonacci(i), end=' ')
print()

# Question 22: Factorial using recursion
def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial_recursive(n-1)

print("Question 22: Factorial using recursion")
num = int(input("Enter a number for recursive factorial: "))
print("Factorial:", factorial_recursive(num))
print()

# Question 23: Sum of digits using recursion
def sum_digits_recursive(n):
    if n == 0:
        return 0
    else:
        return n % 10 + sum_digits_recursive(n // 10)

print("Question 23: Sum of digits using recursion")
num = int(input("Enter a number for recursive sum of digits: "))
print("Sum of digits:", sum_digits_recursive(num))
print()
````

````python
## 27. Frame class Account with object variable account holder's name, account number and balance. Create multiple objects of Account class and perform withdraw, deposit and check balance operations on that account. Also, keep track of how many accounts you have created by using a class variable. (5)
````

````python
# Question 27: Account class with operations
class Account:
    count = 0  # Class variable
    
    def __init__(self, name, acc_no, balance):
        self.name = name
        self.acc_no = acc_no
        self.balance = balance
        Account.count += 1
    
    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited {amount}. New balance: {self.balance}")
    
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance: {self.balance}")
        else:
            print("Insufficient balance")
    
    def check_balance(self):
        print(f"Balance for {self.name}: {self.balance}")

# Creating objects
print("Question 27: Account class demonstration")
acc1 = Account("Amit", 101, 5000)
acc2 = Account("Sara", 102, 3000)

acc1.deposit(1000)
acc2.withdraw(500)
acc1.check_balance()
acc2.check_balance()
print("Total accounts created:", Account.count)
print()
````
